```python
# Installing XGBoost for high-accuracy forecasting
!pip install xgboost

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.seasonal import seasonal_decompose

# Setting a professional 'HUD' Dark Theme
plt.style.use('dark_background')
```

```python
# 1. Advanced Data: 12 months of sales history
data = {
    'Month': np.arange(1, 13),
    'Sales': [45000, 48000, 47000, 52000, 55000, 53000, 58000, 62000, 75000, 88000, 92000, 110000]
}
df = pd.DataFrame(data)

# 2. Machine Learning Training
X = df[['Month']]
y = df['Sales']

# We use an XGBoost Regressor for 'Extreme' accuracy
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)
model.fit(X, y)

# 3. 3-Month Future Prediction (Months 13, 14, 15)
future = pd.DataFrame({'Month': [13, 14, 15]})
preds = model.predict(future)

print(f"🚀 AI FORECAST: Month 13 Revenue Target: ₹{round(float(preds[0]), 2)}")
```

Macro-Forecasting (Sales Engine)

```python
# Creating a Seasonality Matrix for 4 Products
products = ['Chargers', 'Phone Covers', 'Screen Guards', 'Bluetooth Buds']
months_labels = ['Jan', 'Feb', 'Mar', 'Apr']

# Simulated sales density
density = np.array([
    [100, 150, 250, 120], # Chargers peak in March
    [300, 280, 310, 350], # Covers are stable
    [50, 45, 180, 60],    # Screen Guards peak in March (Bulk buy?)
    [80, 75, 70, 200]     # Buds peak in April (Sale?)
])

plt.figure(figsize=(10, 6))
sns.heatmap(density, annot=True, xticklabels=months_labels, yticklabels=products, cmap="YlOrRd")
plt.title("🔥 VYAPAR AI: PRODUCT DEMAND HEATMAP")
plt.show()
```

Micro-Intelligence (The Heatmap)

```python
# Logic: Safety Stock = 1.645 * Std_Dev (Risk) * Sqrt(Lead Time)
def get_bot_advice(sales_history):
    std_dev = np.std(sales_history)
    safety_stock = 1.64 * std_dev * np.sqrt(7) # 7 days lead time
    return round(safety_stock)

charger_history = [120, 150, 250, 120]
needed_stock = get_bot_advice(charger_history)

print("🤖 VYAPAR BOT ADVICE:")
print(f"Based on volatility, keep at least {needed_stock} extra 'Chargers' in the warehouse.")
print("Strategy: High volatility detected for March. Pre-order 20% more stock.")
```

The "Vyapar Bot" (Strategic Decision)

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. REGIONAL DATA (Example Sales in Lakhs)
regional_data = {
    'State': ['Bihar', 'Maharashtra', 'Uttar Pradesh', 'Karnataka', 'West Bengal'],
    'Sales_Value': [85000, 120000, 95000, 110000, 70000]
}
df_geo = pd.DataFrame(regional_data)

# 2. CALCULATE PERCENTAGE SHARE
total_sales = df_geo['Sales_Value'].sum()
df_geo['Percentage'] = (df_geo['Sales_Value'] / total_sales) * 100

# 3. VISUALIZATION: PIE CHART
plt.figure(figsize=(10, 7))
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']

plt.pie(df_geo['Sales_Value'],
        labels=df_geo['State'],
        autopct='%1.1f%%',
        startangle=140,
        colors=colors,
        explode=(0.1, 0, 0, 0, 0), # Highlight Bihar as the home state
        shadow=True)

plt.title("📍 Vyapar AI: Regional Market Share (%)", fontsize=14)
plt.show()
```

The Regional Sales Data

```python
# 4. VISUALIZATION: BAR CHART
plt.figure(figsize=(10, 6))
plt.bar(df_geo['State'], df_geo['Sales_Value'], color='skyblue', edgecolor='navy')

plt.title("State-wise Sales Volume Comparison", fontsize=14)
plt.xlabel("States", fontsize=12)
plt.ylabel("Sales (₹)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Comparison Bar Chart (For Accuracy)

```python
top_state = df_geo.loc[df_geo['Sales_Value'].idxmax()]
low_state = df_geo.loc[df_geo['Sales_Value'].idxmin()]

print("🤖 VYAPAR BOT: REGIONAL STRATEGY")
print("-" * 40)
print(f"Dominant Market: {top_state['State']} ({round(top_state['Percentage'], 1)}% share).")
print(f"Growth Opportunity: {low_state['State']} is currently at only {round(low_state['Percentage'], 1)}%.")
print(f"\nAdvice: Your business is strongest in {top_state['State']}. Consider opening a new hub there.")
print(f"Advice: Launch a marketing campaign in {low_state['State']} to increase market penetration.")
```
